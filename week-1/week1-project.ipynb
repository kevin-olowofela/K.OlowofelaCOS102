```python
def simple_interest():
    p=float(input('What is the principal'))
    r=float(input('What is the percentage rate'))
    t=float(input('For how long'))
    print('calculating simple interest.....')
    a=p+(p*r*t/100)
    print(f'Simple interest is {a:.2f}')

def compound_interest():
    p=float(input('What is the principal'))
    r=float(input('What is the percentage rate'))
    t=float(input('For how long'))
    n=float(input('what is the number of times interest is added in a year'))
    print('calculating compound interest......')
    a= p*(1+r/n)**(n*t)
    print(f'Compound interest is {a:.2f}')

def annuity_plan():
    p=float(input('What is the principal'))
    r=float(input('What is the percentage rate'))
    t=float(input('For how long'))
    n=float(input('what is the number of times interest is added in a year'))
    m=float(input('how much do you deposit periodically'))
    print('calculating annuity plan....')
    a=p*m*t*(((1+r/n)**(n*t)-1)/(r/n))
    print(f'Annuity plan is {a:.2f}')

print('1. Simple interest')
print('2. Compound interest')
print('3. Annuity plan')
try:
 ask=int(input('pick a number'))
 if ask==1:
   simple_interest()
 elif ask==2:
   compound_interest()
 elif ask==3:
   annuity_plan()
 else:
   print('Wrong input')
except ValueError:
    print('Please enter a valid number (e.g., 1, 2, or 1000).')
```
