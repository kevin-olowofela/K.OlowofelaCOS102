```python
import numpy as np

print('Welcome to the root calculator')
print('------------------------------')
print('click a number for the equation you want to solve')
print('1. quadratic equation')
print('2. cubic equation')
print('3. quartic equation')
print('4. exit program')

try:
    while True:
        num_input = input('pick a number: ')
        if not num_input.isdigit():
            print('Invalid input. Please enter a number from the menu.')
            continue
            
        num = int(num_input)

        if num == 1:
            A = float(input('what is the first coefficient (A): '))
            B = float(input('what is the second coefficient (B): '))
            C = float(input('what is the third coefficient (C): '))
            coeff = [A, B, C]
            print(f'The roots are: {np.roots(coeff)}\n')
        elif num == 2:
            A = float(input('what is the first coefficient (A): '))
            B = float(input('what is the second coefficient (B): '))
            C = float(input('what is the third coefficient (C): '))
            D = float(input('what is the fourth coefficient (D): '))
            coeff = [A, B, C, D]
            print(f'The roots are: {np.roots(coeff)}\n')
        elif num == 3:
            A = float(input('what is the first coefficient (A): '))
            B = float(input('what is the second coefficient (B): '))
            C = float(input('what is the third coefficient (C): '))
            D = float(input('what is the fourth coefficient (D): '))
            E = float(input('what is the fifth coefficient (E): '))
            coeff = [A, B, C, D, E]
            print(f'The roots are: {np.roots(coeff)}\n')
        elif num == 4:
            print('exiting program')
            break
        else:
            print('invalid choice\n')
            continue
except ValueError:
    print('Error: Invalid input. Please ensure all coefficients are valid numbers.')
except Exception as e:
    print(f'An unexpected error occurred: {e}')

```
