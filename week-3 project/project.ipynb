```python
from tkinter import *
import csv

def welcome(username):
    wind = Toplevel(window)
    wind.title("Access")
    wind.geometry("300x150")
    wind.config(bg="#1e1e1e")
    label_1 = Label(wind, text=f"Welcome\n{username}", font=("Segoe UI", 14, "bold"), fg="#4affff", bg="#1e1e1e")
    label_1.pack(pady=30)

def reject(username):
    wind = Toplevel(window)
    wind.title("Access denied")
    wind.geometry("300x150")
    wind.config(bg="#1e1e1e")
    label_1 = Label(wind, text=f"Sorry {username},\nAccess Denied", font=("Segoe UI", 12), fg="#ff4a4a", bg="#1e1e1e")
    label_1.pack(pady=30)

def enter():
    username=name_entry.get().title()
    department=dept_entry.get().title()
    found=False
    with open ('C:\\Users\\olowo\\Downloads\\GIG-logistics(in).csv','r') as csv_file:
        staff_file=csv.DictReader(csv_file)
        for row in staff_file:
            if row['Name'] == username and row['Department'] == department:
                found = True
                break
    if found:
        welcome(username)
    else:
        reject(username)

window=Tk()
window.geometry("400x450")
window.title("Staff checker")
window.config(bg="#121212") 


Label(window, bg="#121212").pack(pady=10)

layer1=Label(window, text="STAFF NAME", font=("Segoe UI", 10, "bold"), fg="#888888", bg="#121212")
layer1.pack(pady=(10, 0))

name_entry=Entry(window, font=("Segoe UI", 12), fg="white", bg="#2d2d2d", insertbackground="white", borderwidth=0, highlightthickness=1, highlightbackground="#444444")
name_entry.pack(pady=10, ipady=5, padx=50, fill=X)

layer2=Label(window, text="DEPARTMENT", font=("Segoe UI", 10, "bold"), fg="#888888", bg="#121212")
layer2.pack(pady=(10, 0))

dept_entry=Entry(window, font=("Segoe UI", 12), fg="white", bg="#2d2d2d", insertbackground="white", borderwidth=0, highlightthickness=1, highlightbackground="#444444")
dept_entry.pack(pady=10, ipady=5, padx=50, fill=X)


Enter_button=Button(window, text="VERIFY ACCESS", font=("Segoe UI", 10, "bold"), bg="#00adb5", fg="white", activebackground="#007d82", activeforeground="white", bd=0, cursor="hand2", command=enter)
Enter_button.pack(pady=40, ipadx=20, ipady=10)

window.mainloop()
```
